```python
import numpy as np
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, Arc
from matplotlib import ticker
from matplotlib.ticker import FormatStrFormatter
```

```python
url_start = "https://www.basketball-reference.com/leagues/NBA_{}_shooting.html"
years = np.arange(1997, 2025)
```

```python
for year in years:
    url = url_start.format(year)
    data = r.get(url)

    with open('C:\\Users\\Luca\\Documents\\jupyter notebook\\NBA shot location\\Seasons\{}.html'.format(year), "w+") as f:
        data = data.text.encode('ascii', 'ignore')
        f.write(str(data))
```

```python
dfs_shooting = []

for year in years:
    with open("C:\\Users\\Luca\\Documents\\jupyter notebook\\NBA shot location\\Seasons\{}.html".format(year)) as f:
        page = f.read()
    soup = bs(page, "html.parser")
    
    soup.find('tr', class_="over_header").decompose()

    rows = soup.findAll(lambda tag: tag.name == "th" and tag.has_attr('data-stat') and tag['data-stat'] == "ranker")    #rimuovo righe e colonne Rk

    for row in rows:
        row.decompose()


    rows = soup.findAll(lambda tag: (tag.name == "th" or tag.name == "td") and tag.has_attr('data-stat') and tag['data-stat'] == "age")     #rimuovo righe e colonne age

    for row in rows:
        row.decompose()


    rows = soup.findAll(lambda tag: (tag.name == "th" or tag.name == "td") and tag.has_attr('data-stat') and tag['data-stat'] == "awards")

    for row in rows:
        row.decompose()

    shooting_table = soup.find(id="shooting")
    shooting = pd.read_html(StringIO(str(shooting_table)))[0]

    shooting = shooting.fillna(0)

    groups = ['{start}-{finish} Season'.format(start=int(year)-1, finish=year)] * 8 + ['% of FGA by Distance'] * 6 + ['FG% by Distance'] * 6 + ['% of FG Ast\'d'] * 2 + ['Dunks'] * 2 + ['Corner 3s'] * 2 + ['Heaves'] * 2
    subgroups = [
        'Player', 'Team', 'Pos', 'G', 'GS', 'MP', 'FG%', 'Dist.', 
        '2P', '0-3', '3-10', '10-16', '16-3P', '3P', 
        '2P', '0-3', '3-10', '10-16', '16-3P', '3P', 
        '2P', '3P', '%FGA', '#', '%3PA', '3P%', 'Att.', 'Md.'
    ]
    multiindex = [(group, sub) for group, sub in zip(groups, subgroups)]

    shooting.columns = pd.MultiIndex.from_tuples(multiindex)
    
    dfs_shooting.append(shooting)
```

```python
def draw_court(ax=None, color='black', lw=2, outer_lines=False):

    if ax is None:
        ax = plt.gca()

    #rettangolo di gioco
    court = Rectangle((-250,-47.5), 500, 470, linewidth=lw, color=color, fill=False)

    #genero il canestro
    hoop = Circle((0,0), radius=7.5, linewidth=lw, color=color, fill=False)

    #tabellone
    backboard = Rectangle((-30, -7.5), 60, -1, linewidth=lw, color=color)

    #arco del paint
    paintArc = Arc((0, 0), 80, 80, theta1=0, theta2=180, linewidth=lw, color=color)

    #top free throw arc
    top_free_throw_arc = Arc((0, 142.5), 120, 120, theta1=0, theta2=180, linewidth=lw, color=color, fill=False)

    #bottom free throw arc
    bottom_free_throw_arc = Arc((0, 142.5), 120, 120, theta1=180, theta2=0, linewidth=lw, color=color, linestyle='dashed')

    #rettantolo esterno dell'area
    outer_paint_rectangle = Rectangle((-80, -47.5), 160, 190, linewidth=lw, color=color, fill=False)

    #rettangolo interno dell'area
    inner_paint_rectangle = Rectangle((-60, -47.5), 120, 190, linewidth=lw, color=color, fill=False)

    #Arco da 3 punti
    right_corner_3 = Rectangle((-220, -47.5), 0, 140, linewidth=lw, color=color)
    left_corner_3 = Rectangle((220, -47.5), 0, 140, linewidth=lw, color=color)
    three_arc = Arc((0,0), 475, 475, theta1=22, theta2=158, linewidth=lw, color=color)

    #centro campo
    outer_half_court_arc = Arc((0,422.5), 120, 120, theta1=180, theta2=0, linewidth=lw, color=color)
    inner_half_court_arc = Arc((0,422.5), 40, 40, theta1=180, theta2=0, linewidth=lw, color=color)


    court_elements = [court, hoop, backboard, paintArc, top_free_throw_arc, bottom_free_throw_arc, outer_paint_rectangle, inner_paint_rectangle, right_corner_3, left_corner_3, three_arc, outer_half_court_arc, inner_half_court_arc]

    for element in court_elements:
        ax.add_patch(element)


    #coordinate per zone di tiro
    shots = [
        {"zone": "0-3"},
        {"zone": "3-10"},
        {"zone": "10-16"},
        {"zone": "16-3P"},
        {"zone": "3P"}
    ]

    ax.set_yticklabels([])
    ax.set_xticklabels([])
    
    return ax

```

```python
from matplotlib.colors import Normalize
import numpy as np
from scipy.stats import gaussian_kde

dfs_shots_charts = []
seasons = ['1999-00', '2009-10', '2014-15', '2019-20']
vmins = {}
vmaxes = {}

for i in range(len(seasons)):
    dfs_shots_charts.append(pd.read_csv(f"C:/Users/Luca/Documents/jupyter notebook/NBA shot location/Shot_Data/nba_shot_data_{seasons[i]}.csv"))
    dfs_shots_charts[i].fillna(0, inplace=True)

# Estraggo le coordinate LOC_X e LOC_Y dal DataFrame
    x = dfs_shots_charts[i]['LOC_X']
    y = dfs_shots_charts[i]['LOC_Y']

# Calcola la densità con KDE
    kde = gaussian_kde([x, y])  # KDE prende una lista di array
    density = kde([x, y])       # Calcola la densità in corrispondenza delle tue coordinate
    
    vmins[seasons[i]] = np.percentile(density, 1)  # Assicura che vmin sia almeno 1
    vmaxes[seasons[i]] = np.percentile(density, 99)         # Usa il 99° percentile come valore massimo

print("vmins: ",vmins)
print("vmaxes: ",vmaxes)
```

```python
import os
from IPython.display import display, Image
from ipywidgets import Dropdown, widgets
from matplotlib.colors import LogNorm


preloaded_plots = {}

def jointplots(season, dfs_shots_charts):
    
    output_dir = 'C:\\Users\\Luca\\Documents\\jupyter notebook\\NBA shot location\\Preloaded_jointplots'
    os.makedirs(output_dir, exist_ok=True)

    fig = plt.figure(figsize=(6,6))
    grid = plt.GridSpec(4, 4, hspace=0.3, wspace=0.3)


    for season, df in zip(seasons, dfs_shots_charts):
        norm = LogNorm(vmin=vmins[season], vmax=vmaxes[season])

        cmap = plt.cm.plasma

        joint_shot_chart = sns.jointplot(x='LOC_X', y='LOC_Y', data=df, kind='kde', norm=norm, fill=True, thresh=0, cmap=cmap, levels=100)
        joint_shot_chart.figure.set_size_inches(12,11)

        ax = joint_shot_chart.ax_joint

        draw_court(ax, color="white", lw=1)

        plt.xlim(-250,250)
        plt.ylim(422.5, -47.5)

        ax.set_xlabel('')
        ax.set_ylabel('')
        ax.tick_params(labelbottom='off', labelleft='off')
        ax.set_title(f'{season} season')


        #salva il grafico
        file_path = os.path.join(output_dir, f"{season}_shot_chart.png")
        joint_shot_chart.figure.savefig(file_path, dpi=100)

        preloaded_plots[season] = joint_shot_chart.figure
        plt.close(joint_shot_chart.figure)

    return preloaded_plots


preloaded_plots = jointplots(seasons, dfs_shots_charts)


season_selector = widgets.Dropdown(
    options=seasons,
    description='Season:',
    style={'description_width': 'initial'}
)

# Funzione di aggiornamento
def update_plot(season):
    fig = preloaded_plots[season]
    display(fig)

# Widget interattivo
widgets.interactive(update_plot, season=season_selector)

#plt.show()
```

# Quali tipologie di tiri sono più comuni?

```python
fig, ax = plt.subplots()

shots_0_3 = np.empty(shape=(0,27))
shots_3_10 = np.empty(shape=(0,27))
shots_10_16 = np.empty(shape=(0,27))
shots_16_3P = np.empty(shape=(0,27))
shots_3P = np.empty(shape=(0,27))

for i in range(years):
    shots_0_3 = np.append(shots_0_3, dfs_shooting[i]['% of FGA by Distance']['0-3'].sum() / dfs_shooting[i]['% of FGA by Distance']['0-3'].count())
    shots_3_10 = np.append(shots_3_10, dfs_shooting[i]['% of FGA by Distance']['3-10'].sum() / dfs_shooting[i]['% of FGA by Distance']['3-10'].count())
    shots_10_16 = np.append(shots_10_16, dfs_shooting[i]['% of FGA by Distance']['10-16'].sum() / dfs_shooting[i]['% of FGA by Distance']['10-16'].count())
    shots_16_3P = np.append(shots_16_3P, dfs_shooting[i]['% of FGA by Distance']['16-3P'].sum() / dfs_shooting[i]['% of FGA by Distance']['16-3P'].count())
    shots_3P = np.append(shots_3P, dfs_shooting[i]['% of FGA by Distance']['3P'].sum() / dfs_shooting[i]['% of FGA by Distance']['3P'].count())

plt.plot(all_seasons, shots_0_3, 'g.-')
plt.plot(all_seasons, shots_3_10, 'b.-')
plt.plot(all_seasons, shots_10_16, 'y.-')
plt.plot(all_seasons, shots_16_3P, 'r.-')
plt.plot(all_seasons, shots_3P, 'c.-')

ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=0))

for y_value in plt.gca().yaxis.get_majorticklocs():
    plt.axhline(y = y_value, color = 'gray', linestyle = 'dotted', linewidth = 0.5)

plt.title('% di Tiri dal campo sul totale')
plt.legend(['Layups', 'Tiri dal pitturato', 'Tiri da media distanza', 'Tiri da dentro l\'arco', 'Tiri da 3 punti'], bbox_to_anchor=(1.0, 0.99))
plt.show()


    
```

```python

positions = ['PG','SG','SF','PF','C']



data_list = []

for i in range(years):
    sumCounters = {}
    posCounters = {}

    for pos in positions:
        posCounters[pos] = len(dfs_efficiency[i][dfs_efficiency[i]['Pos'] == pos])
        sumCounters[pos] = dfs_efficiency[i].loc[dfs_efficiency[i]['Pos'] == pos].sum() 


    meanTSByPos = {}

    for pos in positions:
        if posCounters[pos] > 0:
            meanTSByPos[pos] = sumCounters[pos]['TS%'] / posCounters[pos]

        else:
            meanTSByPos[pos] = 0

    data = {
        'Year' : all_seasons[i],
        'PG_TS%' : meanTSByPos['PG'],
        'SG_TS%' : meanTSByPos['SG'],
        'SF_TS%' : meanTSByPos['SF'],
        'PF_TS%' : meanTSByPos['PF'],
        'C_TS%' : meanTSByPos['C']
    }

    data_list.append(data)

df_TSpercs = pd.DataFrame(data_list)
df_TSpercs

```

# I giocatori di quali ruoli sono maggiormente efficienti?

```python
from nba_api.stats.endpoints import shotchartdetail, commonteamroster
from nba_api.stats.static import teams
import pandas as pd

# Creare un dizionario con gli ID delle squadre
teams_ids = {team['abbreviation']: team['id'] for team in teams.get_teams()}

# Creare un DataFrame vuoto per accumulare i dati
all_shots = pd.DataFrame()
cols_to_keep = ['PLAYER_NAME', 'TEAM_NAME', 'SHOT_ZONE_AREA', 'SHOT_ZONE_RANGE', 'LOC_X', 'LOC_Y', 'SHOT_MADE_FLAG', 'GAME_DATE']

seasons = ['1999-00', '2009-10', '2014-15', '2019-20']

#iterare sule stagioni 1999-00, 2009-10, 2014-15, 2019-20
for i in range(len(seasons)):
    # Iterare su ciascuna squadra
    for team_abbreviation, team_id in teams_ids.items():
        try:
            # Ottenere il roster della squadra per la stagione specificata
            roster = commonteamroster.CommonTeamRoster(team_id=team_id, season=f"{seasons[i]}")   #senza roaster non funzionava
            players = roster.get_data_frames()[0]  # DataFrame con i giocatori
            
            # Iterare sui giocatori della squadra
            for _, player in players.iterrows():
                try:
                    # Richiedere i dati di tiro del giocatore
                    shotchart = shotchartdetail.ShotChartDetail(
                        team_id=team_id,
                        player_id=player['PLAYER_ID'],  # ID del giocatore
                        season_nullable=f"{seasons[i]}",  # stagione
                        season_type_all_star="Regular Season"  # Regular Season o Playoffs
                    )
                    
                    # Estrarre il DataFrame
                    player_shots = shotchart.get_data_frames() [0]
                    player_shots = player_shots[cols_to_keep]
                    player_shots = player_shots.query('LOC_X != 0 & LOC_Y != 0')

                    # Aggiungere al DataFrame principale
                    all_shots = pd.concat([all_shots, player_shots], ignore_index=True)
                
                except Exception as e:
                    print(f"Errore con il giocatore {player['PLAYER']} (ID: {player['PLAYER_ID']}): {e}")
        
        except Exception as e:
            print(f"Errore con la squadra {team_abbreviation} (ID: {team_id}): {e}")

    all_shots.to_csv(f"C:\\Users\\Luca\\Documents\\jupyter notebook\\NBA shot location\\Shot_Data\\nba_shot_data_{seasons[i]}.csv", index=False)


print(f"Totale tiri raccolti: {len(all_shots)}")
print(all_shots.head())



```
